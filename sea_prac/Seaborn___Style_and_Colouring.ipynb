#**SubPlots :-**
---

```python
fig, axes = plt.subplots(1,2 , figsize= (20,7)) 
fig.suptitle ('2 Plots')
sns.boxplot( ax=axes [0] ,x='species', y="body_mass_g", data=df_penguins,hue='sex')
sns.violinplot(  ax=axes [1]  , x='species', y="body_mass_g", data=df_penguins,hue='sex',split=True )
```

```python
fig, axes = plt.subplots(1,3 , figsize= (20,10)) 
fig.suptitle ('3 Plots')
sns.boxplot( ax=axes [0] ,x='alive', y="age", data=df_titanic,hue='sex')
sns.violinplot(  ax=axes [1]  , x='alive', y="age", data=df_titanic,hue='sex',split=True )
sns.barplot( ax=axes [2]  , x='alive', y="age", data=df_titanic,hue='sex' )

```

```python
fig, axes = plt.subplots(2,2 , figsize= (20,15)) 
fig.suptitle ('4 Plots')
sns.swarmplot( ax=axes [0][0] ,x= 'sex' , y ='total_bill' , data= df_tips , hue= 'time')
sns.boxplot(  ax=axes [0] [1] , x= 'sex', y='total_bill' , data=df_tips, hue= 'day')
sns.stripplot ( ax = axes [1][0] ,x= 'sex' , y ='total_bill' , data= df_tips )
sns.violinplot( ax = axes [1][1], x= 'sex', y='total_bill' , data=df_tips, hue= 'time' , split= True )
```
